#### Check if a given number is between 50 and 60

```python
n = int(input())
if 50<n<60:                   #n>50 and n<60:
    print('Good')
else:
    print('Not good')
```

#### Else if

```python
n = int(input())
if n<5:
    print('less')
elif n==5:
    print('equal')
else:
    print('great')
```

##### Code to check if a given year is leap year

```python
year = int(input())
if (year %4==0 and year%100 != 0) or year%400==0:
    print('Leap year')
else:
    print('Not leap year')
```

#### Ladder 

```python
mark = int(input())
if mark>=0 and mark<=100:
    if mark>=90:
        print('S grade')
    elif mark>=80: #if mark>=80 and mark<90:
        print('A grade')
    elif mark>=70:
        print('B grade')
    elif mark>60:
        print('C grade')
    elif mark==60:
        print('D grade')
    else:
        print('No grade')
else:
    print('Invalid input')
```
