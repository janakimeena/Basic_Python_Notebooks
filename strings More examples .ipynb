#### Find all occurrences of a substring in a given string

```python
s = input()
sub = input()
occ = s.find(sub)
while occ!=-1:
    print(occ)
    occ = s.find(sub, occ+len(sub))
```

#### Reverse a string - from a given position 'l'

```python
def rev_print(s,l):
    if l<0:
        return
    print(s[l])
    rev_print(s,l-1)
rev_print('apple',3)
```

#### 12 hour format to 24 hour format

```python
time = input()
l = time.split()
time_sp = l[0].split(':')
if time_sp[0]=='12':
    if l[1]=="AM":
        time_sp[0]='00'
    else:
        time_sp[0]='12'
else:
    if l[1]=="PM":
        time_sp[0] = str(int(time_sp[0])+12)
res = ':'.join(time_sp)
print(res)
```

#### Count number of vowels and constanants

```python
string = input()
vow = 0
slen = len(string)
'''for ch in string:
    if ch=='a' or ch=='e' or ch=='i' or ch == 'o' or ch=='u':
        vow+=1'''
vowels = 'aeiou'
for ch in string:
    if ch in vowels:
        vow+=1        
for i in range(0,slen):
    if string[i]=='a' or string[i]=='e' or string[i]=='i' or string[i] == 'o' or string[i]=='u':
        vow+=1
print('Number of vowels ',vow)
print('Number of constants ', slen-vow)
```

#### Regular expression check validity of email ID

```python
import re
#email=input()
if re.match('[^\.]*@[a-zA-Z]+.[a-zA-Z]{3}','m#janakimeena@gmail.com'):
    print('Yes')
else:
    print('No')
```

#### Regular expression to check if a given number is an integer

```python
import re
if re.match('^0|[+|-]?[1-9][0-9]*$','+5'):
    print('Yes')
else:
    print('No')
```
