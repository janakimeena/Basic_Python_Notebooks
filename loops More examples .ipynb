#### Stair case pattern

```python
ht = int(input())
for i in range(1,ht+1):
    for j in range(1,i+1):
        print('#',end='')
    print()
```

#### Check if a number is prime - Checking till square root of n

```python
import math
n = int(input())
if n<2:
    print('Invalid input')
else:
    st = 2
    stop = int(math.sqrt(n))
    for i in range(st,stop+1):
        if n%i==0:
            print('Not prime')
            break        
    else:
        print('Prime')
```

#### Sum of digits

```python
n = int(input())
sum_val = 0
while n!=0:
    sum_val+=n%10 #sum_val = sum_val+n%10
    n//=10 # n = n//10
print(sum_val)
```
